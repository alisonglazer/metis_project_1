# Project 1: Using the combined stations and turnstiles df to generate specific dfs for popular weekdays and times of day

## Importing all the things

```python
import numpy as np
import pandas as pd
import pickle
```

```python
with open('data/combined_df.pickle','rb') as read_file:
    combined_df = pickle.load(read_file)
```

```python
combined_df.head()
```

## Pickling all the major times of day for tech hubs and sending to Alison for Mapping

```python
am_weekday_df = combined_df[np.logical_and(combined_df.weekday<=4,np.logical_and(combined_df.hour<=12,combined_df.hour>=9))]

am_weekday_stations = pd.DataFrame(am_weekday_df.groupby(['station','GTFS Latitude', 'GTFS Longitude', 'tech_borough'])['exit_counts'].agg(np.median).reset_index())
am_weekday_stations = am_weekday_stations.sort_values('exit_counts',ascending=False)                                  
#top_am_weekday_stations = am_weekday_stations[0:51]
```

```python
with open('data/am_weekday_stations_combined.pickle', 'wb') as write_file:
    pickle.dump(am_weekday_stations,write_file)
```

```python
pm_weekday_df = combined_df[np.logical_and(combined_df.weekday<=4,np.logical_and(combined_df.hour>=12+6,combined_df.hour<=12+9))]

pm_weekday_stations = pd.DataFrame(pm_weekday_df.groupby(['station', 'GTFS Latitude', 'GTFS Longitude', 'tech_borough'])['exit_counts'].agg(np.median).reset_index())
pm_weekday_stations = pm_weekday_stations.sort_values('exit_counts',ascending=False)
#top_pm_weekday_stations = pm_weekday_stations[0:51]
```

```python
with open('data/pm_weekday_stations_combined_exits.pickle', 'wb') as write_file:
    pickle.dump(pm_weekday_stations, write_file)
```

```python
pm_weekday_df_entry = combined_df[np.logical_and(combined_df.weekday<=4,np.logical_and(combined_df.hour>=12+6,combined_df.hour<=12+9))]

pm_weekday_stations_entry = pd.DataFrame(pm_weekday_df_entry.groupby(['station', 'GTFS Latitude', 'GTFS Longitude', 'tech_borough'])['entry_counts'].agg(np.median).reset_index())
pm_weekday_stations_entry = pm_weekday_stations_entry.sort_values('entry_counts',ascending=False)
```

```python
with open('data/pm_weekday_stations_combined_entry.pickle', 'wb') as write_file:
    pickle.dump(pm_weekday_stations_entry, write_file)
```
