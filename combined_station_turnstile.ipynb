## Import Clean Turnstile DF and Station DF

```python
import pickle
import numpy as np
import pandas as pd
```

```python
with open('data/clean_df.pickle','rb') as read_file:
    clean_df = pickle.load(read_file)
```

```python
def readStationData(link):
    """
    This function reads in data from an online MTA station location dataset into a DataFrame
    ---
    input: link to dataset
    output: DataFrame
    """
    station_data = pd.read_csv(link, header = 0) #names = cols)
    drop_cols = ['Complex ID', 'GTFS Stop ID', 'Line',
                 'Daytime Routes', 'Structure','North Direction Label', 'South Direction Label']
    station_data = station_data.drop(drop_cols, axis=1)
    return station_data
```

```python
station_df = readStationData('http://web.mta.info/developers/data/nyct/subway/Stations.csv')
```

```python
with open('data/station_df.pickle', 'wb') as to_write:
    pickle.dump(station_df, to_write)
```
